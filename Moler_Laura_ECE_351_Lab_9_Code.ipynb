Additionally, plot the phase of the Fourier transform.
```python
#Insert all necessary imports here
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
#Insert your function to perform the fft here
def FastFourierTransform(x, fs):
    '''
    This function performs a Fast Fourier Transform on the given signal (x).
    Given the signal and sampling frequency, the function will return the computed frequencies, magnitude, and phase
    for the given signal.
    INPUTS:
        x - desired signal to go through transform
        fs - desired sampling frequency
    OUTPUTS: **RETURNS IN THIS ORDER**
        freq - computed frequencies given fs
        mag - computed magnitude of the signal
        phi - computed phase of the signal
    '''
    N = len(x) #find length of signal
    xFFT = scipy.fftpack.fft(x) #perform fft
    xFFTShift = scipy.fftpack.fftshift(xFFT) #shift zero frequency components to center of spectrum
    freq = np.arange(-N/2, N/2) * fs / N #compute frequencies for output signal
    mag = np.abs(xFFTShift) / N #compute mags
    phi = np.angle(xFFTShift) #compute phase
    return freq, mag, phi
```

```python
#Insert code for Part 1 Task 1 here
sampleFreq = 100
steps = 1/sampleFreq
t = np.arange(0, 2, steps)
x = np.cos(2*np.pi*t)
freq, mag, phi = FastFourierTransform(x, sampleFreq)

#plotting
plt.figure(tight_layout=True)
ax1 = plt.subplot(3, 1, 1)
ax2 = plt.subplot(3, 2, 3)
ax3 = plt.subplot(3, 2, 4)
ax4 = plt.subplot(3, 2, 5)
ax5 = plt.subplot(3, 2, 6)

#put plots in subplots
ax1.plot(t, x)
ax1.set_xlabel("t [s]")
ax1.set_ylabel("x(t)")

ax2.stem(freq, mag)
ax2.set_ylabel("|x(t)|")
ax2.set_xlabel("frequency")

ax3.stem(freq, mag)
ax3.set_xlim([-2,2])
ax3.set_xlabel("frequency")

ax4.stem(freq, phi)
ax4.set_ylabel("/_ x(t)")
ax4.set_xlabel("frequency")

ax5.stem(freq, phi)
ax5.set_xlim([-2,2])
ax5.set_xlabel("frequency")

#show graphs
plt.suptitle(r"User-Defined FFT of cos(2$\pi$t)")
plt.show()
```

```python
#Insert your amended fft function here
def FastFourierTransformAMEND(x, fs):
    '''
    AMENDMENT: all elements of magnitude <1e-10 get phase = 0
    This function performs a Fast Fourier Transform on the given signal (x).
    Given the signal and sampling frequency, the function will return the computed frequencies, magnitude, and phase
    for the given signal.
    INPUTS:
        x - desired signal to go through transform
        fs - desired sampling frequency
    OUTPUTS: **RETURNS IN THIS ORDER**
        freq - computed frequencies given fs
        mag - computed magnitude of the signal
        phi - computed phase of the signal
    '''
    N = len(x) #find length of signal
    xFFT = scipy.fftpack.fft(x) #perform fft
    xFFTShift = scipy.fftpack.fftshift(xFFT) #shift zero frequency components to center of spectrum
    freq = np.arange(-N/2, N/2) * fs / N #compute frequencies for output signal
    mag = np.abs(xFFTShift) / N #compute mags
    phi = np.angle(xFFTShift) #compute phase
 
    #amendment
    for i in range(len(mag)):
        if np.abs(mag[i]) < 1e-10:
            phi[i] = 0

    return freq, mag, phi
```

5. Run for the Fourier series approximation of the square wave plotted in **Lab 8**, using only the $N = 15$ case through your updated version of the fft developed in **Task 4**. For the time-domain plot, use $0 \le t \le 16$s. Further, use $T = 8$ for the period of the series as you did in **Lab 8**.

```python
#Insert code for Part 1 Task 5 here
def findB(k):
    '''
    This function finds the bk values using the Fourier series formula
    INPUTS: - k : which value in the series the user is searching for
    OUTPUTS:
            - returns bk of the above Fourier series based off the k value given
    '''
    return ((2/(k*np.pi))*(1-np.cos(k*np.pi)))
N = 15
T = 8
sampleFreq = 100
steps = 1/sampleFreq
t = np.arange(0,16, steps)
x = np.zeros_like(t)
for j in range(N+1):
    bk = findB(j+1)
    x += (bk*np.sin((j+1)*(2*np.pi/T)*t))
    
freq, mag, phi = FastFourierTransformAMEND(x, sampleFreq)

#plotting
plt.figure(tight_layout=True)
ax1 = plt.subplot(3, 1, 1)
ax2 = plt.subplot(3, 2, 3)
ax3 = plt.subplot(3, 2, 4)
ax4 = plt.subplot(3, 2, 5)
ax5 = plt.subplot(3, 2, 6)

#put plots in subplots
ax1.plot(t, x)
ax1.set_xlabel("t [s]")
ax1.set_ylabel("x(t)")

ax2.stem(freq, mag)
ax2.set_ylabel("|x(t)|")
ax2.set_xlabel("frequency")

ax3.stem(freq, mag)
ax3.set_xlim([-3,3])
ax3.set_xlabel("frequency")

ax4.stem(freq, phi)
ax4.set_ylabel("/_ x(t)")
ax4.set_xlabel("frequency")

ax5.stem(freq, phi)
ax5.set_xlim([-3,3])
ax5.set_xlabel("frequency")

#show graphs
plt.suptitle(r"User-Defined FFT of Lab 8")
```
