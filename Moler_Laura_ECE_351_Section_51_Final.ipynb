### Question 1 (10 points)

```python
# Insert code for Question 1 here
steps = 1e-3

#configure plots
plt.figure(figsize=(12,12), tight_layout=True)
ax1 = plt.subplot(4, 1, 1)
ax2 = plt.subplot(4, 1, 2)
ax3 = plt.subplot(4, 1, 3)
ax4 = plt.subplot(4, 1, 4)

#plot
t = np.arange(0, 3+steps, steps)
y1 = 4*np.cos(t)
ax1.plot(t, y1, "r", label="4cos(t)")
ax1.grid()
ax1.set_xlabel("t [s]")
ax1.set_ylabel("$x_1$")
ax1.legend()

t = np.arange(0, 2+steps, steps)
y2 = 1.5*np.sin(2*t)
ax2.plot(t, y2, "b", label="1.5sin(2t)")
ax2.grid()
ax2.set_xlabel("t [s]")
ax2.set_ylabel("$x_2$")
ax2.legend()

t = np.arange(0, 5+steps, steps)
y3 = 3*(np.cos(3*t)**2)*np.exp(-t)
ax3.plot(t, y3, "g", label="$3cos^2(3t)e^{-t}$")
ax3.grid()
ax3.set_xlabel("t [s]")
ax3.set_ylabel("$x_3$")
ax3.legend()

t = np.arange(0, 6+steps, steps)
y4 = 0.5*np.exp(-5*t)
ax4.plot(t, y4, color="black", label="$0.5e^{-5t}$")
ax4.grid()
ax4.set_xlabel("t [s]")
ax4.set_ylabel("$x_4$")
ax4.legend()

plt.suptitle("Question 1")
plt.show()
```

### Question 3 (10 points)

```python
# Insert code for Question 3 here
#H(s) = s**2-s+9 / s**4+3s**2+11s

#print zeros and poles (labeled and formatted)
num = [1, -1, 9]
den = [1, 0, 3, 11, 0, 0] #extra 0 added for residue
res, pol, _ = sig.residue(num,den)
print(f"Residue: {res} \nPoles: {pol}")

#plot impulse response (solid red) and step response (dash blue) - 12x8 from 0<t<10s
num = [1, -1, 9]
den = [1, 0, 3, 11, 0] #H(s)
steps = 1e-3
t = np.arange(0, 10+steps, steps)
tout,yout = sig.impulse((num,den), T=t)
toutStep, youtStep = sig.step((num,den), T=t)

#plot
plt.figure(figsize=(12, 8))
plt.plot(tout,yout, "r") #impulse response
plt.plot(toutStep, youtStep, "--b") #step response
plt.grid()
plt.xlabel("t [s]")
plt.ylabel("H(s)")
plt.suptitle("Question 3")
plt.show()
```

>Q: Is either response stable?
>
>A: The system is not stable because two of the poles are in the right-half plane. Therefore, neither response is stable.
